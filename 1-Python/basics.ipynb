```python
## Arithmetic Operations
# Python supports various arithmetic operations such as addition, subtraction, multiplication, division, modulus, exponentiation, and floor division.
a = 15
b = 4
print("Addition:", a + b)            # Addition
print("Subtraction:", a - b)         # Subtraction
print("Multiplication:", a * b)      # Multiplication
print("Division:", a / b)            # Division
print("Modulus:", a % b)             # Modulus
print("Exponentiation:", a ** b)     # Exponentiation
print("Floor Division:", a // b)     # Floor Division
```

```python
## Operator Precedence
# Python follows the standard mathematical operator precedence. Parentheses can be used to change the order of operations.
result = (a + b) * (a - b) / b ** 2
print("Result with operator precedence:", result)
```

```python
### Comparison Operators
# Comparison operators are used to compare two values. They return a Boolean value (True or False) based on the comparison.
x = 10
y = 20
print("Equal to:", x == y)            # Equal to
print("Not equal to:", x != y)        # Not equal to
print("Greater than:", x > y)         # Greater than
print("Less than:", x < y)            # Less than
print("Greater than or equal to:", x >= y)  # Greater than or equal to
print("Less than or equal to:", x <= y)     # Less than or equal to
```

```python
## Logical Operators
# Logical operators are used to combine conditional statements. They return a Boolean value based on the logic applied.
a = True
b = False
print("Logical AND:", a and b)  # Logical AND
print("Logical OR:", a or b)    # Logical OR
print("Logical NOT:", not a)     # Logical NOT
```

```python
## Simple example of calculator 
def calculator(op, x, y):
    if op == 'add':
        return x + y
    elif op == 'subtract':
        return x - y
    elif op == 'multiply':
        return x * y
    elif op == 'divide':
        return x / y
    else:
        return "Invalid operation"
        

# Example usage
print(calculator('add', 10, 5))      # Output: 15
print(calculator('subtract', 10, 5)) # Output: 5
print(calculator('multiply', 10, 5)) # Output: 50
print(calculator('divide', 10, 5))   # Output: 2.0
```

```python
## simple calculator

first_number  = eval(input("Enter the first number"))
second_number  = eval(input("Enter the second number"))
operation = input("Enter the operation that you want to do using this calculator like + , - , * etc")

if operation == "+":
    print(first_number + second_number)
elif operation == "-":
    print(first_number - second_number) 
elif operation == "*":
    print( first_number * second_number)
elif operation == "%":
    print(first_number % second_number) 
elif operation == "//":
    print(first_number // second_number)                
```

## Loops --- for , while do-while

### Loops -----  Loops in Python are used to repeat actions efficiently. The main types are For loops (counting through items) and While loops (based on conditions).


### For loops -------- For Loop  is used to iterate over a sequence such as a list, tuple, string or range. It allow to execute a block of code repeatedly, once for each item in the sequence.

- iterating over a range 
- iterating over a string

### While loops ------ In Python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the line immediately after the loop in the program is executed.

### Loops control statements
- break 
- continue
- pass

### Nested loops

### Practical example and comman errors 


```python
## rang
"""
In Python, the range() function is primarily used within for loops to control the number of iterations.
It generates an immutable sequence of integers "on demand," making it memory efficient, especially for large sequences. """

"""
Syntax and Parameters
The range() function has three possible ways to be called, with one required argument and two optional ones: 
"""

"""
range(stop)
range(start, stop)
range(start, stop, step)

"""

range(5)
```

```python
## for loop 

## This generates a sequence from 0 up to (but not including) 5, using the default step of 1.

for i in range(5):
    print(i)
```

```python
## While loop 

""" The while loop continue to execute until condition become true """
count = 0
while count <5:
    print(count)
    count = count+1
```

```python
count = 0

while count%2==0:
    print(count)
    count = count+1
```

```python
### Loop control statements

## break --- the break statement exit the loop prematurlly

for i in range(10):
    if i==5:
        break
    print(i)


```

```python
## continue in loop 

# the continue statement   skip the current iteration and continue with the next

for i in range(10):
    if i==5:
        continue
    print(i)
```

```python
## Pass
# the pass statement is a null operation it does nothing


for i in range(10):
    if i==3:
        pass
    print(i)
```

```python
## Nested loop 

# nested loop is nothing but a loop inside  the loop

for i in range(5):
    for j in range(3):
        print(f"i:{i} and j:{j}")
```
